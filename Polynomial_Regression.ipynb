```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X= 6* np.random.rand(100,1) - 3
y= 0.5*X**2+1.5*X+2+np.random.randn(100,1)
```

```python
X, y
```

```python
#scatter plot
plt.scatter(X,y,color='g')
plt.xlabel('X dataset')
plt.ylabel('y dataset')
plt.title('X VS y')
plt.show()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
plt.scatter(X_train, y_train)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
model
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, model.predict(X_test))
print("The R2 score for linear regression is: ", score*100, '%')
```

```python
#visualize this linear regression model for training dataset
plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train), 'r')
```

```python
#visualize this linear regression model for testing dataset
plt.scatter(X_test, y_test)
plt.plot(X_test, model.predict(X_test), 'r')
```

Polynomial regression = b0 + b1X + b2X^2 + b3X^3 + ....bnX^n

```python
#Applying polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly= PolynomialFeatures(degree=2) #creates new input columns (x, x, etc.), not a model.
X_train_poly= poly.fit_transform(X_train) #learns + transform the dataset
X_test_poly = poly.transform(X_test)  #only transform the testing data
```

```python
X_train_poly
```

```python
polymodel = LinearRegression()
polymodel
```

We are not changing what were trying to predict  were only changing how we describe the inputs.

```python
polymodel.fit(X_train_poly, y_train)
poly
```

```python
X_train_poly.shape
```

```python
plt.scatter(X_train, y_train, color='g')
plt.scatter(X_train, polymodel.predict(X_train_poly),color='r')
plt.show()
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, polymodel.predict(X_test_poly))
print("The R2 score for polynomial regression is: ", score*100, '%')
```

```python
#prediction on new dataset
X_new = np.linspace(-3, 3, 200).reshape(200,1) #created 200 evenly shaped dataset between -3 to 3 (inclusive). reshape changes it into 2D array
X_new_poly= poly.transform(X_new)
X_new_poly
```

```python
y_new = polymodel.predict(X_new_poly)          # model predictions

plt.scatter(X_train, y_train, color='g', label='Training data')       # actual data
plt.plot(X_new, y_new, color='r', label='Predictions')
plt.legend()
plt.show()
```

Pipelining in programming (and ML) means:

Connecting multiple steps together in a fixed sequence so that data flows from one step to the next automatically.

All the processing steps are defined once, and the machine (pipeline) handles them in order.



```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=2)),
    ('linreg', LinearRegression())
])

pipeline.fit(X_train, y_train)       # automatically fits poly + regression
y_pred = pipeline.predict(X_test)    # automatically transforms X_test and predicts

```
